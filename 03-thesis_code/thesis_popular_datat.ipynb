## NIO

```python
import pandas as pd

import os
from datetime import datetime

nio_df = pd.read_csv("tmp/popularity_export/NIO.csv")
nio_df['timestamp'] = nio_df['timestamp'].map(lambda dt_string: datetime.strptime(dt_string, '%Y-%m-%d %H:%M:%S'))
```

```python
nio_df.shape
```

```python
# The following is based off of https://stackoverflow.com/a/42372823/3833068
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# plot
plt.plot(nio_df['timestamp'], nio_df['users_holding'])
plt.title('Nio (NIO) Popularity over Time')
plt.ylabel('Unique Robinhood Users Holding Shares')
# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.show()
```

```python
nio_df = nio_df[(nio_df['timestamp'] > '2020-07-01 00:00:00') & (nio_df['timestamp'] < '2020-07-15 23:59:00')]
```

```python
nio_df.shape
```

```python
nio_df.head
```

```python
# The following is based off of https://stackoverflow.com/a/42372823/3833068
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# plot
plt.plot(nio_df['timestamp'], nio_df['users_holding'])
plt.title('Nio (NIO) Popularity 01.07.2020 - 15.07.2020')
plt.ylabel('Unique Robinhood Users Holding Shares')
# beautify the x-labels
plt.gcf().autofmt_xdate()

plt.show()
```

```python
nio_df[nio_df['users_holding'] == nio_df['users_holding'].max()]
```
